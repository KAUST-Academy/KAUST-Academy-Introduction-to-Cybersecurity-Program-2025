```python
# This program demonstrates the One-Time Pad cipher encryption and decryption.
import random
import string

# Generate a random key for the One-Time Pad encryption.
def generate_random_key(length):
    return ''.join(random.choice(string.ascii_uppercase) for _ in range(length))

# Encrypt the text using the One-Time Pad.
def one_time_pad_encrypt(plaintext, key):
    encrypted_text = ""
    for i, char in enumerate(plaintext):
        if char.isalpha():
            is_upper = char.isupper()
            char = char.upper()
            key_char = key[i].upper()
            # Apply a shift based on the key character
            encrypted_char = chr(((ord(char) - ord('A') + ord(key_char) - ord('A')) % 26) + ord('A'))
            encrypted_text += encrypted_char if is_upper else encrypted_char.lower()
        else:
            encrypted_text += char
    return encrypted_text

# Decrypt the text using the One-Time Pad.
def one_time_pad_decrypt(ciphertext, key):
    decrypted_text = ""
    for i, char in enumerate(ciphertext):
        if char.isalpha():
            is_upper = char.isupper()
            char = char.upper()
            key_char = key[i].upper()
            # Apply the reverse shift based on the key character
            decrypted_char = chr(((ord(char) - ord('A') - (ord(key_char) - ord('A'))) % 26) + ord('A'))
            decrypted_text += decrypted_char if is_upper else decrypted_char.lower()
        else:
            decrypted_text += char
    return decrypted_text

# Main program to demonstrate One-Time Pad cipher.
def main():
    print("One-Time Pad Cipher Program")

    # Ask the user for input.
    choice = input("Do you want to (E)ncrypt or (D)ecrypt a message? ").strip().upper()
    if choice not in ['E', 'D']:
        print("Invalid choice. Please choose 'E' for encrypt or 'D' for decrypt.")
        return

    text = input("Enter your message: ").strip()
    key = input("Enter your key (or press Enter to generate a random key): ").strip()

    if not key:
        key = generate_random_key(len(text))
        print(f"Generated Key: {key}")

    if len(key) < len(text):
        print("Error: The key must be at least as long as the message.")
        return

    if choice == 'E':
        encrypted_message = one_time_pad_encrypt(text, key)
        print(f"Encrypted message: {encrypted_message}")
    elif choice == 'D':
        decrypted_message = one_time_pad_decrypt(text, key)
        print(f"Decrypted message: {decrypted_message}")

# Entry point of the program.
if __name__ == "__main__":
    main()
```
